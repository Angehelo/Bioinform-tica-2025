#### 6.- Realiza una función que realice una digestión con las siguientes enzimas de restricción en nuestra secuencia. La función tiene que devolver la siguiente información: (2 pto)

2. Enzimas que no tienen sitios de corte (si las hay)
3. Resultado de la digestión: los fragmentos resultantes de nuestra secuencia
4. Indicar número y tamaño de cada fragmento

```python
import re
enzR = {
    'HindIII' : 'A*AGCTT',
    'BamHI' : 'G*GATCC',
    'AluI' : 'AG*CT',
    'Sau3AI' : '*GATC',
    'EcoRI' : 'G*AATTC'
} 
def digestion(secuencia, enzimas): #Creo las estructuras vacias, en el caso de diccionario uso llaves y en listas corchetes
    enzimas_con_corte = {} #uso diccionario porque me interesa conocer la secuencia, no solo el nombre ya que luego tengo que saber donde corta
    enzimas_sin_corte = []

    for enzima, patron in enzimas.items(): #Quitamos asterisco
        patron_sin_asterisco = patron.replace("*", "")
        resultado = re.search(patron_sin_asterisco, secuencia)
        
        if resultado:
            enzimas_con_corte[enzima] = resultado.start()
        else:
            enzimas_sin_corte.append(enzima)

    # Resultados
    print("Enzimas con sitio de corte:")
    for enzima, pos in enzimas_con_corte.items():
        print(f"{enzima}: posición {pos}")

    print("\nEnzimas sin sitio de corte:")
    for enzima in enzimas_sin_corte:
        print(enzima)
```
